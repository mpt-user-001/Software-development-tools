```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("Ogurcov.csv")
df
```

```python
df = df.drop(columns=["Rank"])
for index, row in df.iterrows():
    for i in range(len(df.columns)):
        if row[i] <= 0.0000001:
            row[i]= row[i]*10000000
        if row[i] <= 0.000001:
            row[i]= row[i]*10000000
        if row[i] <= 0.00001:
            row[i]= row[i]*1000000
        if row[i] <= 0.0001:
            row[i]= row[i]*100000
        if row[i] <= 0.001:
            row[i]= row[i]*10000
        if row[i] <= 0.01:
            row[i]= row[i]*1000
        if row[i] <= 0.1:
            row[i] = row[i]*100
        if row[i] <= 1:
            row[i] = row[i]*10
df = df.astype(int)
df
```

```python
rfc = RandomForestClassifier() #случайный лес
knc = KNeighborsClassifier() #K-ближайшие соседи
gnb = GaussianNB() #наивный байесовский классификатор

prc = PolynomialFeatures() #Полиномиальная регрессия
lrc = LinearRegression() #Линейная регрессия
rfr = RandomForestRegressor() #Регрессия Случайного леса
```

```python
Y = df["Genre"]
X = df.drop(columns = "Genre")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "\nTEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]
    
    rfc.fit(X_train_kfold,Y_train_kfold)
    predict_rfc = rfc.predict(X_test_kfold)

    knc.fit(X_train_kfold,Y_train_kfold)
    predict_knc = knc.predict(X_test_kfold)

    gnb.fit(X_train_kfold,Y_train_kfold)
    predict_gnb = gnb.predict(X_test_kfold)
    
    print("\n")
    print("Набор предсказаний Случайный лес: ",predict_rfc)
    print("Набор предсказаний К-ближайших соседей: ",predict_knc)
    print("Набор предсказаний наивного баеса: ",predict_gnb)
    print("Набор предсказаний Случайный лес: ",predict_rfc,"\n")
```

```python
Y = df["Global_Sales"]
X = df.drop(columns = "Global_Sales")

kf = KFold(n_splits=5, shuffle=False, random_state=None)
kf
```

```python
for train_index, test_index in kf.split(df):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train_kfold = X.iloc[train_index]
    X_test_kfold = X.iloc[test_index]
    Y_train_kfold = Y.iloc[train_index]
    Y_test_kfold = Y.iloc[test_index]
    
    #Обучение и предсказание Регрессии случайного леса
    rfr.fit(X_train_kfold,Y_train_kfold)
    predict_rfr = rfr.predict(X_test_kfold)
    
    #Обучение и предсказание Линейной регрессии
    lrc.fit(X_train_kfold,Y_train_kfold)
    predict_lrc = lrc.predict(X_test_kfold)

    #Обучение и предсказание Полиномиальная регрессия
    lprc = LinearRegression()
    X_train_kfold_prc = prc.fit_transform(X_train_kfold)
    lprc.fit(X_train_kfold_prc,Y_train_kfold)
    X_test_kfold_prc = prc.fit_transform(X_test_kfold)
    predict_prc = lprc.predict(X_test_kfold_prc)
    
    print("\n")
    print("Набор предсказаний регрессии случайного леса: ",predict_rfr)
    print("Набор предсказаний линейной регрессии: ",predict_lrc)
    print("Набор предсказаний полиномиальной регрессии: ", predict_prc,"\n")
```
